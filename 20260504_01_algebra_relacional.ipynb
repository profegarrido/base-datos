# Álgebra relacional

**Conjunto de operaciones** que se utilizan para manipular relaciones.

### Tipo de operaciones
<ul>
    <li><b>Unión, Diferencia e Intersección</b>. Igual que en conjunto, pero aplicado a relaciones.</li>
    <li><b>Selección</b>. Selecciona ciertas tuplas de una relación.</li>
    <li><b>Proyección</b>. Selecciona ciertas columnas de una relación.</li>
    <li><b>Productos y Joins</b>. Composición de relaciones.</li>
    <li><b>Renombrado</b> de relaciones y atributos.</li>
    <li><b>División, agrupación y ordenación</b> de relaciones.</li>
    <li><b>Modificación de tuplas</b>. Insertar, actualizar y eliminar</li>
</ul>

### Selección ($\sigma$)
<p>Permite filtrar filas (tuplas) que cumplen con una condición específica.</p>
<p>Ejemplo: Empleados que tengan un sueldo mayor o igual a 8000 fecha de ingreso sea a partir del uno de enero del 2003.</p>



$$\sigma_{sueldo \ge 8000 \land ingreso \ge 01/01/2003}(Empleado)$$

#### Intersección ($\cap$)

$$r \cap s = R -(R - S)$$
Se Trata de una relación con las tuplas que están en R y S también.
Operación válida entre relaciones compatibles.
Renglones duplicados se eliminan

$$r = \pi_{sueldo}(\sigma_{depto = A1}(Empleado))$$
$$s = \pi_{sueldo}(\sigma_{depto = A2}(Empleado))$$
$$r \cap s$$

### Join Externo 

$$ R =\bowtie_{\theta} S \quad  R \Join=_{\theta} S \quad R =\bowtie=_{\theta} S$$

<ul>
    <li>Las tuplas que no tienen una tupla relacionada se eliminan del resultado.</li>
    <li>Las tuplas que tienen valor nulo en los atributos del JOIN también se eliminan.</li>
    <li>El Join externo, por el contrario, permite conservar en el resultado todas las tuplas que estén en R, en S o en ambas, ya sea que tengan o no tuplas coincidentes en la otra relación.</li>
    <li>Esto satisface la necesidad de las consultas donde las tuplas de las dos tablas se van a combinar para emparejar las correspondientes filas, pero aquellas tuplas originalmente perdían por no tener valores coincidentes con tuplas de la otra relación, ahora estarán presentes.</li>
</ul>
<ul>
    <li>La operación de join externo por la izquierda (=⋈θ​) conserva todas las tuplas de la primera relación R (o relación de la izquierda) en $R =\bowtie_{\theta} S$, si no se encuentra una tupla coincidente en S. Los atributos de S del resultado se "rellenan" con valores nulos.</li>
    <li>El join externo por la derecha (⋈=θ​), conserva en el resultado de R⋈=θ​S todas las tuplas de la relación S (la de la derecha).</li>
    <li>El join externo completo (=⋈=θ​), conserva todas las tuplas de ambas relaciones, izquierda y derecha, cuando no se encuentran tuplas coincidentes, rellenándolas con valores nulos si es necesario.    </li>
</ul>

### Ordenar ($\tau$)
$$\tau_{atributo1 \ [asc \ | \ desc], \ atributo2 \ [asc \ | \ desc], \ ...} (R)$$
<ul>
    <li>Este operador devuelve la relación R, pero en el orden expresado por la secuencia de atributos.</li>
    <li>El operador solo tiene sentido cuando se usa como el operador final de una expresión.</li>
</ul>

### Mantenimiento de datos

<ul>
    <li>Borrado. Tiene la notación R=R−e.</li>
        <ul>
            <li>Donde <b>e</b> puede ser una tupla constante especificada o el resultado de una consulta en álgebra relacional.</li>
            <li>Restricciones: Tupla con el mismo grado que la relación R y con los valores en los dominios adecuados.</li>
        </ul>
    <li>Inserción. Tiene la notación R=R$\cup$e.</li>
        <ul>
            <li>Donde <b>e</b> puede ser una tupla constante especificada o el resultado de una consulta en álgebra relacional.</li>
            <li>Restricciones: Tupla con el mismo grado que la relación R y con los valores en los dominios adecuados.</li>
        </ul>
    <li>Actualización. BORRADO + INSERCIÓN.</li>
        <ul>
            <li>Se suele utilizar el operador de proyección generalizado, donde en la lista de proyección aparecen nombres de atributos (si no se modificaron) o bien éstos involucrados en operaciones aritméticas.</li>
        </ul>
</ul>
