<p><strong>SQLite </strong> es un sistema de gestión de bases de datos relacionales ligero, rápido y de código abierto. A diferencia de bases de datos tradicionales como MySQL o SQL Server, no requiere configurar un servidor independiente; los datos se almacenan directamente en un único archivo en el disco, lo que facilita su integración.</p>

### Características Principales
<ul>
    <li><b>Sin servidor (Serverless)</b>: No necesita procesos en segundo plano ni instalación de servicios. La base de datos es leída y escrita directamente por la aplicación que la utiliza.</li>
    <li><b>Autónoma y portátil</b>: Toda la base de datos reside en un solo archivo (generalmente con extensión .db o .sqlite) que se puede copiar o transferir fácilmente entre distintos dispositivos.</li>
    <li><b>Totalmente compatible (ACID)</b>: Cumple con las propiedades de atomicidad, consistencia, aislamiento y durabilidad, lo que garantiza que las transacciones de datos sean seguras y fiables.</li>
    <li><b>Ligera</b>: Es extremadamente pequeña en tamaño (alrededor de unos pocos cientos de kilobytes) y consume muy poca memoria.</li>
</ul>

### Cuándo se utiliza
<p>Gracias a su practicidad, es el motor de bases de datos más implementado a nivel mundial. Sus usos más comunes incluyen:</p>
<ul>
    <li><b>Aplicaciones móviles</b>: Es el motor estándar en Android e iOS para guardar historiales, configuraciones y datos de apps.</li>
    <li><b>Desarrollo de escritorio</b>: Utilizado por programas como Adobe Photoshop Lightroom, navegadores web (para guardar marcadores o caché) y otros programas de software.</li>
    <li><b>Dispositivos IoT y embebidos</b>: Ideal para equipos con recursos de hardware limitados, como electrodomésticos inteligentes o dispositivos médicos.</li>
    <li><b>Desarrollo web y prototipado</b>: Muy útil para sitios web de bajo tráfico o entornos de desarrollo donde no se justifica el costo ni la complejidad de un servidor de base de datos completo.</li>
</ul>

### Limitaciones
<p>Aunque es excelente para uso local y monousuario, no es recomendable si necesitas:</p>
<ul>
    <li><b>Concurrencia masiva</b>: No está diseñado para soportar cientos o miles de usuarios escribiendo en la base de datos al mismo tiempo.</li>
    <li><b>Escalabilidad horizontal o replicación:</b> Carece de herramientas nativas para distribuir la base de datos entre múltiples servidores (servidores espejo).</li>
    <li><b>Sistemas de permisos complejos:</b> Ofrece un control de usuarios y privilegios muy limitado en comparación con bases de datos empresariales.</li>
</ul>
      

````python
import sqlite3

# 1. Conectar a la base de datos (en memoria para esta clase)
conexion = sqlite3.connect(':memory:')

# 2. Crear el cursor (nuestro mensajero para ejecutar comandos SQL)
cursor = conexion.cursor()

print("Conexión establecida con éxito. El motor de SQLite está listo.")
````

### <b>CRUD</b> 
<p>Es un acrónimo fundamental en el mundo de la programación y la gestión de bases de datos que describe las cuatro operaciones básicas para manipular información:</p>
<ul>
    <li><b>C (Create - Crear):</b> Permite añadir o insertar nueva información en un sistema (por ejemplo, registrarse en una web).</li>
    <li><b>R (Read - Leer):</b> Permite consultar o recuperar los datos existentes (por ejemplo, buscar un contacto en tu agenda o ver tu perfil).</li>
    <li><b>U (Update - Actualizar):</b> Permite modificar o editar datos que ya están guardados (por ejemplo, cambiar la contraseña de una cuenta).</li>
    <li><b>D (Delete - Borrar):</b Permite eliminar de forma permanente o temporal un registro (por ejemplo, cancelar una suscripción o borrar una foto).></li>
</ul>

<p>Casi cualquier aplicación que se utilice a diario (como plataformas de e-commerce, redes sociales o aplicaciones de notas) se construye sobre estos cuatro cimientos.</p>

<div style="padding: 15px; border-left: 6px solid #d9534f; background-color: #fdf7f7; color: #d9534f;">
    <strong>Regla de oro de seguridad:</strong> Nunca concatenar strings de Python para rellenar datos en consultas SQL. Utilizar placeholders (marcador de posición) (?) para sanitizar los datos automáticamente y evitar ataques de Inyección SQL.
</div>


````python
# C (Create) - Crear la tabla del catálogo de productos/servicios
cursor.execute('''
    CREATE TABLE IF NOT EXISTS productos (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        nombre TEXT NOT NULL,
        categoria TEXT,
        precio REAL
    )
''')
conexion.commit() # Confirmamos la estructura en la memoria
print("Tabla 'productos' creada.")
````

````python
# C (Create/Insert) - Inserción de datos masiva de forma segura usando executemany
servicios_iniciales = [
    ('Cambio de Aceite', 'Mantención', 45000),
    ('Alineación', 'Mantención', 25000),
    ('Scanner Diagnóstico', 'Electrónica', 30000),
    ('Frenos Completos', 'Mecánica', 85000)
]
````

````python
# El "?" actúa como marcador de posición seguro
cursor.executemany('INSERT INTO productos (nombre, categoria, precio) VALUES (?, ?, ?)', servicios_iniciales)
conexion.commit()
print(f"Se han insertado {cursor.rowcount} registros iniciales.")
````

````python
# R (Read) - Consultas usando SELECT, WHERE y ORDER BY
print("--- EJECUTANDO SELECT + WHERE + ORDER BY ---")

query_filtro = '''
    SELECT nombre, precio 
    FROM productos 
    WHERE categoria = ? AND precio >= ?
    ORDER BY precio DESC
'''
````

````python
# Pasamos los parámetros de filtrado en una tupla ordenada
cursor.execute(query_filtro, ('Mantención', 20000))

# Recuperamos todos los registros que coincidieron
resultados = cursor.fetchall()
for fila in resultados:
    print(f"Servicio: {fila[0]} | Precio: ${fila[1]:,.0f}")
````

````python
# U (Update) - Modificar un registro existente
cursor.execute('UPDATE productos SET precio = ? WHERE nombre = ?', (48000, 'Cambio de Aceite'))
conexion.commit()
````

````python
# D (Delete) - Eliminar un registro específico
# NOTA: Al pasar un solo parámetro en la tupla, es obligatorio incluir la coma final (valor,)
cursor.execute('DELETE FROM productos WHERE nombre = ?', ('Scanner Diagnóstico',))
conexion.commit()
````

````python
print("Actualización y eliminación completadas con éxito.")
````

### Mapeo Dinámico desde Diccionarios y CSV
<p>En entornos de ingeniería de datos, las estructuras de las tablas o los datos de origen no siempre están fijos. En esta sección aprenderemos a automatizar la creación de tablas desde meta-información y a poblar la base de datos masivamente desde un archivo plano CSV.</p>


````python
# Generación dinámica de esquemas DDL mediante un Diccionario de Python
estructura_cliente = {
    "rut": "TEXT PRIMARY KEY",
    "nombre": "TEXT NOT NULL",
    "telefono": "TEXT",
    "edad": "INTEGER"
}
````

````python
# Unimos las llaves y valores en un string compatible con SQL
columnas_sql = ", ".join([f"{columna} {tipo}" for columna, tipo in estructura_cliente.items()])
sql_dinamico = f"CREATE TABLE IF NOT EXISTS clientes ({columnas_sql})"

print("Sentencia SQL autogenerada:")
print(sql_dinamico)

cursor.execute(sql_dinamico)
conexion.commit()
````

### Buenas Prácticas
<p>Errores comunes observados en estudiantes:</p>

<ul>
    <li><b>Olvidar el .commit(): </b>Si ejecutas INSERT, UPDATE o DELETE y no confirmas la transacción, los datos no se persistirán.</li>
    <li><b>Error de Sintaxis en Tuplas Unitarias: </b>Escribir (valor) en vez de (valor,) al pasar un solo parámetro al cursor.</li>
    <li><b>No cerrar conexiones: </b><Siempre es una buena práctica liberar los recursos del sistema al terminar los scripts./li>
</ul>

````python
# Cierre limpio de la sesión de base de datos
conexion.close()
print("Conexión cerrada. Recursos liberados correctamente.")
````
